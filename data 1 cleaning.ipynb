Using the melt function to get the Segment row values (Consumer, Consumer Total, Corporate, etc.) and the Ship Mode row values (First Class, Same Day, etc.) and putting them in as column values where the header is Segment and Ship Mode respectively and putting their values as Sales.

```python
data1 = pd.melt(data, id_vars = [('Segment>>', 'Ship Mode>>')],var_name=['Segment','Ship Mode'],value_name='Sales')
data1
```

Renamed all my column.

```python
data1.columns = ['Order ID', 'Segment', 'Ship Mode', 'Sales']
data1
```

Dropped any NaN values

```python
data1.dropna(subset = ['Sales'], axis = 0, inplace = True)
```

```python
data1.Segment.value_counts()
```

Made a list called segment_vals of the values I wanted to keep in the Segment column.

Got rid of the __ Total Segments since they are just totals.

```python
segment_vals = ['Consumer', 'Corporate', 'Home Office']
data1 = data1.loc[data1['Segment'].isin(segment_vals)]
data1
```

```python
data1.Segment.value_counts()
```

```python
data1.set_index('Order ID', inplace = True)
```

```python
data1
```

```python
with pd.ExcelWriter('clean_data.xlsx') as excel_writer:
    data1.to_excel(excel_writer, sheet_name='Sheet1', index = False)
```
