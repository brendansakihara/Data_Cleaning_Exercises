```python
date_data = pd.read_excel('data_3.xlsx', sheet_name = 'dates')
date_data
```

```python
combine_data = pd.merge(data, date_data, how = 'left', on='Order ID')
```

```python
combine_data
```

```python
clean_data = combine_data.iloc[:,[0,4,1,2,3]]
```

```python
clean_data.set_index('Order ID', inplace = True)
clean_data
```

```python
with pd.ExcelWriter('clean data.xlsx') as excel_writer:
    clean_data.to_excel(excel_writer, sheet_name = 'Sheet1', index = True)
```
