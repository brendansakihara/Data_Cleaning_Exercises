```python
import pandas as pd
import numpy as np
import numbers
```

Imported the dataset and made 'Description' the index.

```python
data = pd.read_excel('data_6.xlsx')
data.set_index('Description', inplace = True)
```

Checked the datatypes and saw that they are all objects.

```python
data.dtypes
```

Took a look at the dataset and saw that some values are letters instead of numbers (like o's instead of 0's, i's instead of 1's).

```python
data.head(15)
```

Renamed all the columns taking out the underscores.

```python
data.columns = ['Basic Unit', 'Beginning Balance', 'Quantity Received', 'Quantity Dispensed', 'Losses and Adjustment', 'Ending Balance (Physical Count)']
data
```

Looked through excel filter's and saw that o's and O's were used as 0's, i's and I's were used as 1's, and 's' and 'S' were used as 5's (decided to use 5 since 5 looked the most like s/S).

```python
data['Beginning Balance'].replace('o', '0', regex=True, inplace = True)
data['Beginning Balance'].replace('O', '0', regex=True, inplace = True)
data['Beginning Balance'].replace('i', '1', regex=True, inplace = True)
data['Beginning Balance'].replace('I', '1', regex=True, inplace = True)
data['Beginning Balance'].replace('s', '5', regex=True, inplace = True)
data['Beginning Balance'].replace('S', '5', regex=True, inplace = True)
```

```python
data['Quantity Received'].replace('o', '0', regex=True, inplace = True)
data['Quantity Received'].replace('O', '0', regex=True, inplace = True)
data['Quantity Received'].replace('i', '1', regex=True, inplace = True)
data['Quantity Received'].replace('I', '1', regex=True, inplace = True)
data['Quantity Received'].replace('s', '5', regex=True, inplace = True)
data['Quantity Received'].replace('S', '5', regex=True, inplace = True)
```

```python
data['Quantity Dispensed'].replace('o', '0', regex=True, inplace = True)
data['Quantity Dispensed'].replace('O', '0', regex=True, inplace = True)
data['Quantity Dispensed'].replace('i', '1', regex=True, inplace = True)
data['Quantity Dispensed'].replace('I', '1', regex=True, inplace = True)
data['Quantity Dispensed'].replace('s', '5', regex=True, inplace = True)
data['Quantity Dispensed'].replace('S', '5', regex=True, inplace = True)
```

```python
data['Losses and Adjustment'].replace('o', '0', regex=True, inplace = True)
data['Losses and Adjustment'].replace('O', '0', regex=True, inplace = True)
data['Losses and Adjustment'].replace('i', '1', regex=True, inplace = True)
data['Losses and Adjustment'].replace('I', '1', regex=True, inplace = True)
data['Losses and Adjustment'].replace('s', '5', regex=True, inplace = True)
data['Losses and Adjustment'].replace('S', '5', regex=True, inplace = True)
```

```python
data['Ending Balance (Physical Count)'].replace('o', '0', regex=True, inplace = True)
data['Ending Balance (Physical Count)'].replace('O', '0', regex=True, inplace = True)
data['Ending Balance (Physical Count)'].replace('i', '1', regex=True, inplace = True)
data['Ending Balance (Physical Count)'].replace('I', '1', regex=True, inplace = True)
data['Ending Balance (Physical Count)'].replace('s', '5', regex=True, inplace = True)
data['Ending Balance (Physical Count)'].replace('S', '5', regex=True, inplace = True)
```

Checked the dataset to make sure the changed went through.

```python
data.head(15)
```

Changed the data type to numeric.

```python
data['Beginning Balance'] = pd.to_numeric(data['Beginning Balance'])
data['Quantity Received'] = pd.to_numeric(data['Quantity Received'])
data['Quantity Dispensed'] = pd.to_numeric(data['Quantity Dispensed'])
data['Losses and Adjustment'] = pd.to_numeric(data['Losses and Adjustment'])
data['Ending Balance (Physical Count)'] = pd.to_numeric(data['Ending Balance (Physical Count)'])
```

Created a row called 'Total' were its shows the sum of each column.

```python
data.loc['Total'] = data.sum(numeric_only= True, axis = 0)
data
```
